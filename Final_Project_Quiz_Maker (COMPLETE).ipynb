```python
#OS is used to clear the program if neccessary/ if user chooses to.
import os
# Sys is used to exit the system if need be.
import sys
# Time is used to create the 10 Minute Timer If need be.
import time
# Random is used for selecting either 10 to 20 questions at random. 
import random
# csv is used for managing cvs files
import csv


def ask_questions(question_list):
    """
    Prints out questions and answers while asking for the user's answer,
    then checks it with the correct answer and increases the score if they got it right.
    Takes a list of dictionaries containing questions, options, and correct answers as an argument -- Hunter coded this
    """
    score = 0
    num_questions = len(question_list)
    for i, question_dict in enumerate(question_list):
        print(f"\nQuestion {i + 1}/{num_questions}: {question_dict['question']}")
        for j, option in enumerate(question_dict['options']):
            print(f"{j + 1}. {option}")
        while True:
            user_answer = input("Your answer (1, 2, or 3): ")
            if user_answer.isdigit() and user_answer in ['1', '2', '3']:
                user_answer = int(user_answer)
                if user_answer == 1:
                    user_answer = "A"
                    break
                elif user_answer == 2:
                    user_answer = "B"
                    break
                else: 
                    user_answer = "C"
                    break
            else:
                print("Invalid input! Please enter 1, 2, or 3.")
        if user_answer == question_dict['correct_answer']:
            print("Correct!")
            score += 1
        else:
            print(f"Wrong! The correct answer is {question_dict['correct_answer']}")
    print(f"\nYou scored {score} out of {num_questions}.\n")
            
def read_questions_from_csv(filename):
    """
    Reads questions from a CSV file and returns a list of questions. 
    Takes the filename of the CSV file containing questions as an argument -- Hunter coded this
    """
    questions = []
    with open(filename, 'r', newline='') as file:
        reader = csv.reader(file)
        for row in reader:
            questions.append(row)
    return questions

def TenQs():
    """
    Randomly selects 10 questions from the CSV file and returns them as a list - Hunter coded this
    """
    questions = read_questions_from_csv('TestBank.csv.csv')
    ten_questions_list = []
    while len(ten_questions_list) < 10:
        selected_question = random.choice(questions)
        question_dict = {
            'question': selected_question[0],
            'options': [selected_question[1], selected_question[2], selected_question[3]],
            'correct_answer': selected_question[4]
        }
        if question_dict not in ten_questions_list:
            ten_questions_list.append(question_dict)
    return ten_questions_list

def TwentyQs():
    """
    Randomly selects 20 questions from the CSV file and returns them as a list -- Hunter coded this
    """
    questions = read_questions_from_csv('TestBank.csv.csv')
    twenty_questions_list = []
    while len(twenty_questions_list) < 20:
        selected_question = random.choice(questions)
        question_dict = {
            'question': selected_question[0],
            'options': [selected_question[1], selected_question[2], selected_question[3]],
            'correct_answer': selected_question[4]
        }
        if question_dict not in twenty_questions_list:
            twenty_questions_list.append(question_dict)
    return twenty_questions_list

def TenQ_Test():
    """
    Creates a test of 10 questions and asks the user to answer them -- Hunter coded this
    """
    try:
        ten_questions = TenQs()
        ask_questions(ten_questions)
    except ValueError as e:
        print(e)
        
def TwentyQ_Test():
    """
    Creates a test of 20 questions and asks the user to answer them -- Hunter coded this
    """
    try:
        twenty_questions = TwentyQs()
        ask_questions(twenty_questions)
    except ValueError as e:
        print(e)

def userCheck():
    """
    Gets the first name, last name, and ID of the user and validates their ID -- Evan coded this
    """
    while True:
        try:
            First_name = input("Enter Your First Name: ")
            print()
            if First_name.isdigit():
                raise TypeError()
            break
            
        except TypeError:
            print("Invalid First Name. Please Try Again")
            print()
          
    while True:
        try:
            Last_name = input("Enter Your Last Name: ")
            print()
            if Last_name.isdigit():
                raise TypeError()
            break
            
        except TypeError:
            print("Invalid Last Name. Please Try Again.")
            print()
            
    while True:
            for i in range(3,-1,-1):
                searchTerm = "0"
                try:
                    if i == 0:
                        print("Number of Attempts Exceeded. Goodbye.")
                        sys.exit()

                    print("Numbers of Attempts Remaining: " + str(i))
                    
                    ID = input("Enter Your ID: ")
                    if len(ID) > 6:
                        raise ValueError
                        continue
                    if ID.isdigit():
                        raise e
                        continue
                    if ID[0] != 'A':
                        raise TypeError()
                        continue
                    if searchTerm in ID[ : ]:
                        raise TypeError()
                        continue
                        
                except TypeError:
                    print("Numbers can only be between 1-9. Please Try Again")
                    print()
                
                except ValueError:
                    print("Too Many Characters. Please Try Again.")
                    print()

                except TypeError:
                    print("First Character Is Not 'A'. Please Try Again")
                    print()

                except Exception as e:
                    print("There are Only Digits. Please Try Again")
                    print()
                else:
                    return

    return


#Asks For The User To Input How Many Questions They Wish To Answer. Only 10 And 20 Are Valid Numbers
#Evan Has Coded This Function
def userQuestionCount():
    while True:
        try:
            userQCount = int(input("Would you like to do 10 Questions? Or 20 Questions? "))

            if userQCount == 10: 
                TenQ_Test() 
                print() 
                break 
            if userQCount == 20:
                TwentyQ_Test()
                print()
                break
            if userQCount != 10 and userQCount != 20:
                raise ValueError()

            # Check elapsed time after each question set
            elapsed_time = time.time() - start_time
            if elapsed_time >= 600:  # 10 minutes (600 seconds)
                print("Time's up! Quiz terminated.")
                sys.exit()

        except ValueError: #Evan Has Coded This
            print("Invalid Answer. Please Try Again.") #Evan Has Coded This
            print() #Evan Has Coded This

#Evan Has Coded This Function
def userCheckClearQuit():
    userCheckClearQuit = input("Do You Want To Clear Or Quit The Program? (Clear = S/ Quit = Q)")
        
    if userCheckClearQuit.upper() == "Q":
        print("Quiting Program. Bye!")
        sys.exit()
        
    if userCheckClearQuit.upper() == "S":
        print("Clearing Program.")
        os.system('cls')
        print()
        main()
        
        
        
            
# Where the User Will Decide How Many Questions They Will Answer.
# I Use the Try/Except to Catch Invalid Numbers/Answers to the "userQCount" variable.
"""
Main function where the quiz is started -- Evan coded this
"""
def main():
    userCheck() #Evan Has Coded This

    # Track start time
    start_time = time.time()

    userQuestionCount() #Evan Has Coded This
            
    userCheckClearQuit() #Evan Has Coded This
        

# Calls the "main" function
if __name__ == "__main__":
    main()
```
