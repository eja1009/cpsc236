## 7.2 Wizard Inventory
Create a program that keeps track of the items that a wizard can carry.

### Console:
```powershell

COMMAND MENU
walk - Walk down the path 
show - Show all items

You can't carry any more items. Drop something first.

Command: show
1.	a wooden staff
2.	a scroll of invisibility
3.	a crossbow
4.	a scroll of uncursing

Command: drop Number: 3

Command: exit

```

### Specifications:
- Use the attached text file named `wizard_all_items.txt` that contains a list of all the items that a wizard can carry. (in the corresponding directory)
- When the user selects the walk command, the program should read the items from the file, randomly pick one, and give the user the option to grab it.
- Your program should create another file that stores the items that the wizard is carrying (`wizard_inventory.txt`). Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry **four** items at a time.
- For the drop command, display an error message if the user enters an invalid number for the item.



````python
### CODE HERE ###
import random
import os

def read_items(filename):
    if os.path.exists(filename):
        with open(filename, 'r') as file:
            return [line.strip() for line in file]
    return []

def write_items(filename, items):
    with open(filename, 'w') as file:
        for item in items:
            file.write(item + '\n')

def show_inventory(inventory):
    if not inventory:
        print("Inventory is empty.")
    else:
        print("Inventory:")
        for i, item in enumerate(inventory, 1):
            print(str(i) + ". " + str(item))

def walk(all_items, wizard_inventory, wizard_inventory_file):
    if len(wizard_inventory) >= 4:
        print("You can't carry any more items. Drop something first.")
    else:
        random_item = random.choice(all_items)
        print("While walking down a path, you see " + str(random_item) + ".")
        grab = input("Do you want to grab it? (y/n): ").lower()
        if grab == "y":
            wizard_inventory.append(random_item)
            print("You picked up " + str(random_item) + ".")
            write_items(wizard_inventory_file, wizard_inventory)

def drop(wizard_inventory, wizard_inventory_file):
    show_inventory(wizard_inventory)
    if wizard_inventory:
        drop_num = input("Number: ")
        if drop_num.isdigit():
            drop_num = int(drop_num)
            if 1 <= drop_num <= len(wizard_inventory):
                dropped_item = wizard_inventory.pop(drop_num - 1)
                print("You dropped " + str(dropped_item) + ".")
                write_items(wizard_inventory_file, wizard_inventory)
            else:
                print("Invalid number. Please enter a valid number.")
        else:
            print("Invalid input. Please enter a number.")
    else:
        print("Inventory is empty.")

def main():
    wizard_inventory_file = "wizard_inventory.txt"
    all_items_file = "wizard_all_items.txt"

    with open(wizard_inventory_file, 'r') as file:
        wizard_inventory = [line.strip() for line in file]

    with open(all_items_file, 'r') as file:
        all_items = [line.strip() for line in file]

    print("The Wizard Inventory Program")

    print("\nCOMMAND MENU")
    print("walk - Walk down the path")
    print("show - Show all items")
    print("drop - Drop an item")
    print("exit - Exit program")
    print()

    while True:
        command = input("Command: ").lower()

        if command == "walk":
            walk(all_items, wizard_inventory, wizard_inventory_file)
            print()
        elif command == "show":
            show_inventory(wizard_inventory)
            print()
        elif command == "drop":
            drop(wizard_inventory, wizard_inventory_file)
            print()
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
````

## 7.3 Monthly Sales
Create a program that reads the sales for 12 months from a file and calculates the total yearly sales as well as the average monthly sales. In addition, this program should let the user edit the sales for any month.

### Console:

### Specifications:
- Use the attached text file named CSV file named `monthly_sales.csv` that contains the month and sales data shown above.
- For the edit command, display an error message if the user doesnâ€™t enter a valid three-letter abbreviation for the month.
- When the user edits the sales amount for a month, the data should be saved to the CSV file immediately. That way, no data is lost, even if the program crashes later.
- Round the results of the monthly average to a maximum of 2 decimal digits.

````python
### CODE HERE ###
import csv

def load_sales_data(filename):
    sales_data = {}
    with open(filename, newline='') as file:
        reader = csv.reader(file)
        header = next(reader) 
        for row in reader:
            month, sales = row
            sales_data[month] = int(sales)
    return sales_data

def save_sales_data(filename, sales_data):
    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Month', 'Sales'])
        for month, sales in sales_data.items():
            writer.writerow([month, sales])

def view_monthly_sales(sales_data):
    print("Monthly Sales")
    for month, sales in sales_data.items():
        print(str(month), "-", str(sales))

def view_yearly_summary(sales_data):
    yearly_total = sum(sales_data.values())
    monthly_average = yearly_total / len(sales_data)
    print("Yearly total:   " + str(yearly_total))
    print("Monthly average:    " + str(round(monthly_average, 2)))

def edit_sales(sales_data):
    month_abbr = input("Three-letter Month: ").capitalize()
    if month_abbr in sales_data:
        new_sales = input("Sales Amount: ")
        sales_data[month_abbr] = int(new_sales)
        save_sales_data("monthly_sales.csv", sales_data)
        print("Sales amount for " + str(month_abbr) + " was modified.")
    else:
        print("Invalid month abbreviation.")

def main():
    sales_data = load_sales_data("monthly_sales.csv")

    
    print("\nCOMMAND MENU")
    print("monthly - View monthly sales")
    print("yearly  - View yearly summary")
    print("edit    - Edit sales for a month")
    print("exit    - Exit program")

    while True:
        command = input("Command: ").lower()

        if command == "monthly":
            view_monthly_sales(sales_data)
            print()
        elif command == "yearly":
            view_yearly_summary(sales_data)
            print()
        elif command == "edit":
            edit_sales(sales_data)
            print()
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()

````
