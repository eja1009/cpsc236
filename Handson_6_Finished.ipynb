## 12.1 Champion Counter
Create a program that reads a text file that contains a list of FIFA World Cup champions and determines the country that has won the most championships.

### Console:
```powershell
Country	        Wins	Years	
=======	        ====	=====	
Argentina       2       1978, 1986
Brazil	        5       1958, 1962, 1970, 1994, 2002
England	        1       1966		
France	        1       1998		
Germany	        4       1954, 1974, 1990, 2014
Italy	        4       1934, 1938, 1982, 2006
Spain	        1       2010
Uruguay	        2       1930, 1950
```

### Specifications:
- Use the attached text file named `world_cup_champions.txt` that contains data like this: 
```powershell
Year,Country,Coach,Captain 
1930,Uruguay,Alberto Suppici,Jos√© Nasazzi 
1934,Italy,Vittorio Pozzo,Gianpiero Combi 
1938,Italy,Vittorio Pozzo,Giuseppe Meazza
...
```
- When the program starts, it should read the text file and use a dictionary to store the required data using the name of each country that has won the World Cup as the key.
- The program should compile the data shown above and display the countries alphabetically.



````python
### CODE HERE ###
def read_world_cup_data(filename):
    world_cup_data = {}
    with open(filename, 'r') as file:
        next(file)  
        for line in file:
            year, country, _, _ = line.strip().split(',')
            if country not in world_cup_data:
                world_cup_data[country] = {'Wins': 1, 'Years': [year]}
            else:
                world_cup_data[country]['Wins'] += 1
                world_cup_data[country]['Years'].append(year)
    return world_cup_data

def display_world_cup_data(world_cup_data):
    print("Country         Wins    Years   ")
    print("=======         ====    =====   ")
    for country in sorted(world_cup_data.keys()):
        wins = world_cup_data[country]['Wins']
        years = ', '.join(world_cup_data[country]['Years'])
        print("{:<15} {:<7}".format(country, wins) + str(years))

def main():
    filename = 'world_cup_champions.txt'
    world_cup_data = read_world_cup_data(filename)
    display_world_cup_data(world_cup_data)

if __name__ == "__main__":
    main()
````

## 12.2 Monthly Sales
Create a program that allows you to view and edit the sales amounts for each month of the current year.

### Console:
```powershell

COMMAND MENU

edit	- Edit sales for specified month 
totals - View sales summary for year 
exit	- Exit program

Command: view

Command: edit

Sales Amount: 15293
Sales amount for Jan is 15,293.00.

Command: totals
Yearly total: 67,855.00
Monthly average: 0

Command: view
Three-letter Month: july 
Invalid three-letter month.

Command: exit Bye!
```

### Specifications:
- Use the attached text file named `monthly_sales.txt` that consists of rows that contain three-letter abbreviations for the month and the monthly sales.
- The program should read the file and store the sales data for each month in a dictionary with the month abbreviation as the key for each item.
- Whenever the sales data is edited, the program should write the changed data to the text file.




````python
def read_monthly_sales(filename):
    monthly_sales = {}
    with open(filename, 'r') as file:
        for line in file:
            month, sales = line.strip().split('\t')
            monthly_sales[month.lower()] = float(sales)
    return monthly_sales

def write_monthly_sales(filename, monthly_sales):
    with open(filename, 'w') as file:
        for month, sales in monthly_sales.items():
            file.write(str(month.capitalize()) + "\t" + str(round(sales, 2)) + "\n")

def view_sales(monthly_sales, month):
    if month.lower() in monthly_sales:
        print("Sales amount for " + str(month.capitalize()) + " is ${:,.2f}".format(float(monthly_sales[month.lower()])) + ".")
    else:
        print("Invalid three-letter month.")

def edit_sales(monthly_sales, month, amount):
    if month.lower() in monthly_sales:
        monthly_sales[month.lower()] = amount
        write_monthly_sales('monthly_sales.txt', monthly_sales)
        print("Sales amount for " + str(month.capitalize()) + " is ${:,.2f}".format(float(amount)) + ".")
    else:
        print("Invalid three-letter month.")

def view_totals(monthly_sales):
    yearly_total = sum(monthly_sales.values())
    monthly_average = yearly_total / len(monthly_sales)
    print("Yearly total: ${:,.2f}".format(float(yearly_total)))
    print("Monthly average: " + "${:,.2f}".format(float(monthly_average)))

def main():
    filename = 'monthly_sales.txt'
    monthly_sales = read_monthly_sales(filename)
    
    
    print("\nMonthly Sales program\n")
    print("COMMAND MENU")
    print("view    - View sales for specified month")
    print("edit    - Edit sales for specified month")
    print("totals  - View sales summary for year")
    print("exit    - Exit program")

    while True:
        command = input("\nCommand: ").lower()

        if command == "view":
            month = input("Three-letter Month: ").lower()
            view_sales(monthly_sales, month)
        elif command == "edit":
            month = input("Three-letter Month: ").lower()
            amount = float(input("Sales Amount: "))
            edit_sales(monthly_sales, month, amount)
        elif command == "totals":
            view_totals(monthly_sales)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
````
