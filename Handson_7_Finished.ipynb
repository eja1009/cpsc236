## 8.1 Tip Calculator
Add exception handling to a Tip Calculator program.

### Console:
```powershell
Tip Calculator

INPUT
Cost of meal: ten
Must be valid decimal number. Please try again. 
Cost of meal: -10
Must be greater than 0. Please try again. 
Cost of meal: 52.31
Tip percent:	17.5
Must be valid integer. Please try again. 
Tip percent:	20

OUTPUT
Cost of meal:   52.31 
Tip percent:    20%
Tip amount:     10.46
Total amount:   62.77
```

### Specifications:
- The program should accept **decimal** entries like 52.31 and 15.5 for the cost of the meal.
- The program should accept **integer** entries like 15, 20, 25 for the tip percent.
- The program should validate both user entries. That way, the user can’t crash the program by entering invalid data.
- The program should only accept numbers that are **greater than 0**.
- The program should round results to a maximum of two decimal places.


````python
### CODE HERE ###
def calculate_tip_and_total(cost_of_meal, tip_percent):
    tip_amount = round(cost_of_meal * (tip_percent / 100), 2)
    total_amount = round(cost_of_meal + tip_amount, 2)
    return tip_amount, total_amount

print("Tip Calculator")
print()

while True:
    try:
        user_meal = float(input("Cost of meal: "))
        if user_meal <= 0:
            raise ValueError("Must be greater than 0. Please try again.")
    except ValueError as ve:
        print("Must be valid decimal number. Please try again.")
        continue
        
    break
    
while True:
    try:
        user_tip = input("Tip percent: ").strip()
        if not user_tip.isdigit():
            raise ValueError("Must be valid integer. Please try again.")
        user_tip = int(user_tip)
        if user_tip not in [15, 20, 25]:
            raise ValueError("Must be 15, 20, or 25. Please try again.")
    except ValueError as ve:
        print(ve)
        continue
    
    break

tip_amount, total_amount = calculate_tip_and_total(user_meal, user_tip)
formatted_tip_percent = "{:d}%".format(user_tip)
formatted_cost_of_meal = "${:.2f}".format(user_meal)
formatted_tip_amount = "${:.2f}".format(tip_amount)
formatted_total_amount = "${:.2f}".format(total_amount)
print()

print("OUTPUT")
print("Cost of meal:  ", formatted_cost_of_meal)
print("Tip percent:   ", formatted_tip_percent)
print("Tip amount:    ", formatted_tip_amount)
print("Total amount:  ", formatted_total_amount)
````

## 8.2 Wizard Inventory

Add exception handling to a program that keeps track of the inventory of items that a wizard can carry. If you still have Wizard Inventory from Handson_5, you can add the exception handling to that program. Otherwise, you can start this program from scratch.

### Console if the program can't find the **inventory** file:

### The error message if the program can’t find the **items** file:
```powershell

```

### Specifications:
- This program should read the text file named `wizard_all_items.txt` that contains all the items a wizard can carry.
- When the user selects the walk command, the program should randomly pick one of the items that were read from the text file and give the user the option to grab it.
- The current items that the wizard is carrying should be saved in an inventory file. Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry four items at a time. For the drop command, display an error message if the user enters an invalid integer or an integer that doesn’t correspond with an item.
- Handle all exceptions that might occur so that the user can’t cause the program to crash. If the all items file is missing, display an appropriate error message and exit the program.
- If the inventory file is missing, display an appropriate error message and continue with an empty inventory for the user. That way, the program will write a new inventory file when the user adds items to the inventory.

````python
### CODE HERE ###
import random
import os
import sys

def read_items(filename):
    try:
        with open(filename, 'r') as file:
            return [line.strip() for line in file]
    except FileNotFoundError:
        if filename == 'wizard_inventory.txt':
            print("Could not find inventory file!")
            print("Wizard is starting with no inventory.")
        if filename == 'wizard_all_item.txt':
            print("Could not find item file!")
            print("Exiting program. Bye!")
            sys.exit()
        
        return []

def write_items(filename, items):
    with open(filename, 'w') as file:
        for item in items:
            file.write(item + '\n')

def show_inventory(inventory):
    if not inventory:
        print("Inventory is empty.")
    else:
        print("Inventory:")
        for i, item in enumerate(inventory, 1):
            print(str(i) + ". " + str(item))

def walk(all_items, wizard_inventory, wizard_inventory_file):
    if len(wizard_inventory) >= 4:
        print("You can't carry any more items. Drop something first.")
    else:
        random_item = random.choice(all_items)
        print("While walking down a path, you see " + str(random_item) + ".")
        grab = input("Do you want to grab it? (y/n): ").lower()
        if grab == "y":
            wizard_inventory.append(random_item)
            print("You picked up " + str(random_item) + ".")
            write_items(wizard_inventory_file, wizard_inventory)

def drop(wizard_inventory, wizard_inventory_file):
    show_inventory(wizard_inventory)
    if wizard_inventory:
        drop_num = input("Number: ")
        try:
            drop_num = int(drop_num)
            if 1 <= drop_num <= len(wizard_inventory):
                dropped_item = wizard_inventory.pop(drop_num - 1)
                print("You dropped " + str(dropped_item) + ".")
                write_items(wizard_inventory_file, wizard_inventory)
            else:
                print("Invalid number. Please enter a valid number.")
        except ValueError:
            print("Invalid item number.")
    else:
        print("Inventory is empty.")

def main():
    wizard_inventory_file = "wizard_inventory.txt"
    all_items_file = "wizard_all_items.txt"

    wizard_inventory = read_items(wizard_inventory_file)
    all_items = read_items(all_items_file)

    print("The Wizard Inventory Program")

    print("\nCOMMAND MENU")
    print("walk - Walk down the path")
    print("show - Show all items")
    print("drop - Drop an item")
    print("exit - Exit program")
    print()

    while True:
        command = input("Command: ").lower()

        if command == "walk":
            walk(all_items, wizard_inventory, wizard_inventory_file)
            print()
        elif command == "show":
            show_inventory(wizard_inventory)
            print()
        elif command == "drop":
            drop(wizard_inventory, wizard_inventory_file)
            print()
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
````
